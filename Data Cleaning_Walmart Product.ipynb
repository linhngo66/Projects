```python
dataset = pd.read_csv("C:/Users/Admin/Documents/Python/marketing_sample_for_walmart_com-product_details__20200101_20200331__30k_data.csv")
dataset

#30000 rows, 14 columns
```

```python
dataset.info()

#Item number, Category, Brand
#Package size and Postal Code drop
```

```python
dataset_mod = dataset.drop(columns=["Package Size","Postal Code","Crawl Timestamp"])
```

### Dropping rows with same Item Number (on-going)

```python
item_no = dataset_mod["Item Number"].value_counts().reset_index()
same_item = list(item_no["index"][item_no["Item Number"]==2])
```

```python
#Get the rows we want to drop because of duplicated Item Number
drop_list = dataset_mod[dataset_mod["Item Number"].isin(same_item)].drop_duplicates("Item Number",keep="last")
drop_list
```

```python
dataset_itemdrop = dataset_mod.drop(drop_list.index)
dataset_itemdrop
```

#### Now that Item Number has no use anymore, we drop this column

```python
dataset_itemdrop.drop(columns="Item Number", inplace = True)
```

### Cut description

```python
dataset_itemdrop["Description"] = dataset_itemdrop["Description"].str.split("|",expand = True)[1]
dataset_itemdrop
```

### Split category

```python
dataset_split = dataset_itemdrop.copy()
```

```python
dataset_split[["CategoryL1","CategoryL2","CategoryL3","CategoryL4","CategoryL5","CategoryL6","CategoryL7","CategoryL8"]] = dataset_split["Category"].str.split("|",expand = True)
```

```python
dataset_split[["CategoryL1","CategoryL2","CategoryL3","CategoryL4","CategoryL5","CategoryL6","CategoryL7","CategoryL8"]].info()
```

```python
dataset_split.drop(columns=["Category","CategoryL5","CategoryL6","CategoryL7","CategoryL8"], inplace=True)
```

### Filling in those with null CategoryL1 (on-going)

```python
dataset_cat = dataset_split.copy()
```

```python
null_L1 = dataset_cat[dataset_cat["CategoryL1"].isnull()]
list(null_L1["Brand"])
```

```python
for x in list(null_L1["Brand"]):
    temp_ds = dataset_cat[dataset_cat["Brand"] == x]
    
    if len(temp_ds) > 1:
        temp_dict = temp_ds["CategoryL1"].value_counts().to_dict()
        index = null_L1[null_L1["Brand"] == x].index
        dataset_cat.loc[(index,"CategoryL1")] = max(temp_dict, key = temp_dict.get)
    else:
        continue

dataset_cat[dataset_cat["CategoryL1"].isnull()]
```

#### Since there are over 500 items unbranded, and data can not be interpreted elsewhere from the dataset, we keep these items as they are

### Calculate %discount

```python
dataset_cat["%discount"] = (1 - (dataset_cat["Sale Price"]/dataset_cat["List Price"]))*100
dataset_cat[dataset_cat["%discount"]!=0]
```

```python
dataset_cat["%discount"].fillna(0, inplace = True)
```
