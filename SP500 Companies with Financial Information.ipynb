# Data analysis and visualization

The following analysis will consider 3 angles: Market Capitalization, Profitability and Market Valuation. The analysis focuses specifically on the higher Sector level, while individual firms are mentioned occasionally.

## Market captilization (market cap)

### Market capitalization by Sector

Market cap of a sector is calculated as the sum of market cap of all companies in the corresponding sector. 

```python
seccap = df.groupby(["Sector"]).sum().sort_values("Market Cap", ascending=False)
seccap = seccap[["Market Cap"]]
seccap.reset_index(inplace=True) 
seccap
```

```python
plt.figure(figsize=(20,10))
plt.bar(seccap["Sector"],seccap["Market Cap"])
plt.xlabel("Sector", fontsize=15)
plt.ylabel("Market Capitalization", fontsize=15)
plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15)
plt.title("Market Capitalization by Sector", fontsize=20) 
```

### Numbers of companies in each sector

```python
df["Sector"].value_counts().sort_values(ascending=False)
```

### Comments

- The biggest sector in terms of market cap is Information Technology (IT) while Consumer Discreationary ranks 1st in population.
- Individual market cap of IT firms is expected to be substantial since IT's market cap doubles that of Consumer Discreationary while the number of IT firms in the list is only slightly lower than Consumer Discreationary.
- Invidual market cap of Healthcare firms is also expected to be substantial give that its market cap is higher than Consumer Discreationary while the population only ranks 5th.

## Profitability 

Profitability is measured using EBITDA margin, which is the percentage of EBITDA in Revenue. Higher EBITDA margin implies lower operating costs, hence, higher profitability. 

The given Earnings per Share is not used to measure profitability as it is subject to individual firms' size, hence, not applicable to compare across firms and across sector. 

```python
#EBITDA margin (in %)
df["EBITDA/Sales"] = df["EBITDA"]/(df["Price"]/df["Price/Sales"])*100
```

### The most profitable sector

```python
plt.figure(figsize=(20,10))
sns.boxplot(df["Sector"],df["EBITDA/Sales"])
plt.xlabel("Sector", fontsize=15)
plt.ylabel("EBITDA as % of Sales", fontsize=15)
plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15)
plt.title("Profitability by Sector", fontsize=20) 
```

### Top 20 most and least profitable companies

**Top 20 most profitable companies, divided by sector**

```python
top20 = df.sort_values("EBITDA/Sales", ascending = False).head(20)
top20["Sector"].value_counts()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(top20["Name"],top20["EBITDA/Sales"],hue="Sector",data=top20, dodge=False).legend(fontsize = 15)
plt.xlabel("Company", fontsize=15)
plt.ylabel("EBITDA as % of Sales", fontsize=15)
plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15)
plt.title("Top 20 most profitable companies", fontsize=20) 
```

**Top 20 least profitable companies, divided by sector**

```python
least20 = df.sort_values("EBITDA/Sales").head(20)
least20["Sector"].value_counts()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(least20["Name"],least20["EBITDA/Sales"],hue="Sector",data=least20, dodge=False).legend(fontsize = 15)
plt.xlabel("Company", fontsize=15)
plt.ylabel("EBITDA as % of Sales", fontsize=15)
plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15)
plt.title("Top 20 least profitable companies", fontsize=20) 
```

**Number of companies with negative earnings, divided by sector**

**Comments**
- The most profitable sector is Telecommunication Service with 2/3 companies in the top 20 most profitable.
- IT has the most companies in both the top 20 most profitable list and the list of companies with negative earnings. This means there might be a big gap between the profit leader and other players in the sector.
- While Financials is big in market cap, the sector is among the least profitable as half of the top 20 least profitable companies are populated by Financials companies.

This can be explained by the sector's heavy capital nature that require firms to take loan and spend a substantial amount of operating incomes (EBITDA) for interest and loan payment.

## Market valuation

This section looks at how the market values different sectors and whether the sectors are undervalued or overvalued.

### The highest valued sector

```python
plt.figure(figsize=(20,10))
sns.boxplot(df["Sector"],df["Price/Earnings"])
plt.xlabel("Sector", fontsize=15)
plt.ylabel("Price/Earnings", fontsize=15)
plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15)
plt.title("Price-to-Earnings Distribution by Sector", fontsize=20) 
```

Since the P/E ratios of most sector, except for Energy, are normally distributed, the mean P/E is used to compare across sector. The median P/E is applied to Energy because of the skewed distribution.

```python
#Setting sector P/Es to their corresponding means
secPE = df.groupby(["Sector"]).mean()
secPE = secPE[["Price/Earnings"]]
secPE.reset_index(inplace=True)
secPE

#Change the mean of Energy into its median
secPE.loc[secPE["Sector"]=="Energy","Price/Earnings"] = df[df["Sector"]=="Energy"].median()["Price/Earnings"]

secPE = secPE.sort_values("Price/Earnings", ascending=False)
```

```python
plt.figure(figsize=(25,10))
plt.bar(secPE["Sector"],secPE["Price/Earnings"])
plt.xlabel("Sector", fontsize=15)
plt.ylabel("Price/Earnings", fontsize=15)
plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15)
plt.title("Price-to-Earnings by Sector", fontsize=20) 
```

### Top 20 companies with the highest P/E

```python
topPE = df[["Name","Sector","Price/Earnings"]].sort_values("Price/Earnings", ascending=False).head(20)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(topPE["Name"],topPE["Price/Earnings"],hue="Sector",data=topPE, dodge=False).legend(fontsize = 15)
plt.xlabel("Company", fontsize=15)
plt.ylabel("Price-to-Earnings", fontsize=15)
plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15)
plt.title("Top 20 highest P/E ratios", fontsize=20) 
```

```python
topPE["Sector"].value_counts()
```

**Comments**

- Despite being the least profitable sector, Energy ranks 2nd in average sector P/E ratio and has the most companies appearing in the top 20 highest P/E list. This means the market is either overvaluing or having high expectation on the future performance of the sector.
- Telecommunication Services seem to be the most undervalued sector since despite high profitability, its P/E is comparable to other sectors and there is no Telecommunication Services company appearing in the top 20 highest P/E list.
